## Univariate Analysis

```python
#Analysing segment of people(customers) buying from the superstore

df["Segment"].value_counts()
```

```python
people_buying_per = round((df["Segment"].value_counts()/len(df["Segment"]))*100,2)
print(people_buying_per)
```

```python
people_buying_per.plot(kind='bar',figsize=(14,6))
```

```python
#Analysing Ship Mode
df["Ship Mode"].value_counts()
```

```python
ship_mode_per = round((df["Ship Mode"].value_counts()/len(df["Ship Mode"]))*100,2)
print(ship_mode_per)
```

```python
ship_mode_per.plot(kind="bar",figsize=(14,6))
```

## Bivariate Analysis

```python
fig,ax = plt.subplots()
colors = {"Consumer":"red","Corporate":"blue","Home Office":"green"}
ax.scatter(df["Sales"],df["Profit"],c=df["Segment"].apply(lambda x:colors[x]))
plt.show()
```

From the above plot, we can say that there is more profit in consumer segment

```python
df.pivot_table(values="Sales",index="Segment",columns="Discount",aggfunc='median')
```

```python
df.pivot_table(values="Profit",index="Segment",columns="Discount",aggfunc='median')
```

From above tables we can say that

For discount more than 30%, the superstore was going on loss, for discount between 0 to 20%, the sales of superstores were average. 

```python
df.loc[(df["Segment"]=="Consumer") & (df["Discount"]==0.1)]["Profit"].plot(kind="hist",bins=50,figsize=(14,6))
```

```python
df.loc[(df["Segment"]=="Consumer") & (df["Discount"]==0.2)]["Profit"].plot(kind="hist",bins=50,figsize=(14,6))
```

```python
df.loc[(df["Segment"]=="Consumer") & (df["Discount"]==0.8)]["Profit"].plot(kind="hist",bins=50,figsize=(14,6))
```

when the superstore is offering discount less than 50%: It's gaining profit\
when when the superstore is offering discount more than 50% (eg 80%): It's facing loss


```python
df.loc[(df["Category"]=="Furniture") & (df["Discount"]==0.2)]["Profit"].plot(kind="hist",bins=50,figsize=(14,6))
```

```python
df.loc[(df["Category"]=="Technology") & (df["Discount"]<=0.3)]["Profit"].plot(kind="hist",bins=50,figsize=(14,6))
```

```python
df.loc[(df["Category"]=="Technology") & (df["Discount"]>=0.3)]["Profit"].plot(kind="hist",bins=50,figsize=(14,6))
```

```python
df.loc[(df["Category"]=="Office Supplies") & (df["Discount"]<=0.3)]["Profit"].plot(kind="hist",bins=50,figsize=(14,6))
```

```python
df.loc[(df["Category"]=="Office Supplies") & (df["Discount"]>=0.3)]["Profit"].plot(kind="hist",bins=50,figsize=(14,6))
```

From the above graphs we can conclude that:

For all the categories: "Office Supplies", "Technology" and "Furniture":
1) When discount is less than 30% in items, sales was going into profit\
2) When discount is greater than 30% in items the Superstore is facing loss

```python
df.groupby(["Segment","Discount"])["Profit"].median().plot(kind="bar",stacked=True,figsize=(14,6))
```

Profit of all the segments when discount was offered by the Superstore.
